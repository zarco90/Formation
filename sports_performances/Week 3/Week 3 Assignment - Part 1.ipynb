## Part 1 - NBA Player Comparision

```python
# As usual, we begin by importing the packages we will need

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# The data consists of the shot log for the NBA season 2016/17

shot = pd.read_csv("Assignment Data/Week 3/NBA shotlog_16_17.csv")
pd.set_option('display.max_columns', 100)
print(shot.columns.tolist())
shot.describe()
```

```python
# A simple plot of coordinates

shot['halfcourt_x'] = np.where(shot['location_x'] < 933/2, 933 - shot['location_x'],shot['location_x'])
shot['halfcourt_y'] = np.where(shot['location_x'] < 933/2, 500 - shot['location_y'],shot['location_y'])
```

```python
hx = shot['halfcourt_x']
hy = shot['halfcourt_y']
plt.figure(figsize=(94/12,50/6))
plt.scatter(hx,hy, s=.01,c='r', marker= '.')
plt.minorticks_on()
plt.grid(which='major', linestyle='-', linewidth='.5', color='black')
plt.grid(which='minor', linestyle=':', linewidth='0.5', color='red')
plt.title("Shots", fontsize = 15)
```

## Westbrook - Durant Comparision

```python
shot_durant = shot.loc[shot['shoot_player'] == 'Kevin Durant']
hxD = shot_durant['halfcourt_x']
hyD = shot_durant['halfcourt_y']

shot_howard = shot.loc[shot['shoot_player'] == 'Dwight Howard']
hxH = shot_howard['halfcourt_x']
hyH = shot_howard['halfcourt_y']

shot_jordan = shot.loc[shot['shoot_player'] == 'DeAndre Jordan']
hxJ = shot_jordan['halfcourt_x']
hyJ = shot_jordan['halfcourt_y']

shot_west = shot.loc[shot['shoot_player'] == 'Russell Westbrook']
hxW = shot_west['halfcourt_x']
hyW = shot_west['halfcourt_y']
```

```python
f = plt.figure(figsize=(94/6,50/6))
ax = f.add_subplot(121)
colors = np.where(shot_west['current_shot_outcome']=='SCORED','r',np.where(shot_west['current_shot_outcome']=='MISSED','b','g'))
ax = plt.scatter(hxW, hyW, s=10, c= colors, marker= '.')
plt.grid(True)
plt.title("Russel Westbrook", fontsize = 15)
plt.xlim(500,950)
plt.ylim(0,500)

ax = f.add_subplot(122)
colors = np.where(shot_durant['current_shot_outcome']=='SCORED','r',np.where(shot_durant['current_shot_outcome']=='MISSED','b','g'))
ax = plt.scatter(hxD, hyD, s=10, c= colors, marker= '.')
plt.grid(True)
plt.title("Kevin Durant", fontsize = 15)
plt.xlim(500,950)
plt.ylim(0,500)
```
